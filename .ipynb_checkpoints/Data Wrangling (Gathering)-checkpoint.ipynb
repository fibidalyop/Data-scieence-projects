# Topic: tweepy

```python
import tweepy as tw
from tweepy import OAuthHandler
```

```python
with open('tweepy_key.json') as f:
    key = js.load(f)
```

```python

consumer_key = key["API Key"]
consumer_secret = key["API Key Secret"]
access_token = key["Access Token"]
access_secret = key["Access Token Secret"]

auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tw.API(auth)
```

```python
import time

sleeptime = 4
pages = tw.Cursor(api.followers).pages()


name = []
verified = []
location = []
followers = []

while True:
    try:
        page = next(pages)
        time.sleep(sleeptime)
    except tw.TweepError: #taking extra care of the "rate limit exceeded"
        time.sleep(60*15) 
        page = next(pages)
    except StopIteration:
        break
    for user in page:
        name.append(user.screen_name)
        followers.append(user.followers_count)
        verified.append(user.verified)
        location.append(user.location)
     
        print(user.screen_name)
```

```python
f= {"name":name,"verified":verified,"location":location,"followers":followers}

friends = pd.DataFrame(data=f)
```
